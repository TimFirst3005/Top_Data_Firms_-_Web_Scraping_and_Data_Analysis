```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
import plotly.express as px
import plotly.offline as po
po.init_notebook_mode(connected=True)
```

```python
url = 'https://www.goodfirms.co/big-data-analytics/data-analytics'
access = r.get(url)
```

```python
content_page = access.content
bs = BeautifulSoup(content_page, 'lxml')
```

```python
tag_name = bs.find_all('span', {'itemprop': "name"})
```

```python
progress_val = bs.find_all('div', {'class': "circle-progress-value"})
```

```python
def extract_detail(tag_lst):
    lst = [tag.text for tag in tag_lst]
    return pd.Series(lst)
```

```python
name = extract_detail(tag_name)
```

```python
service_pct = [percent for percent in enumerate(progress_val)]
service_pct
```

```python
def extract_progress_values(tag_lst):    
    service_pct = [percent[1].text for percent in enumerate(tag_lst) if percent[0]%2 ==0]
    platform_pct = [percent[1].text for percent in enumerate(tag_lst) if percent[0]%2 ==1]       
    return pd.Series(service_pct), pd.Series(platform_pct) 
```

```python
ser, pct = extract_progress_values(progress_val)
```

```python
firm_reviews = bs.find_all('span', {'class': "listinv_review_label"})
reviews = extract_detail(firm_reviews)
reviews
```

```python
values = reviews.apply(lambda x: x.split()[1].strip("("))
values
```

```python
firm_locations = bs.find_all('div', {'class': "firm-location"})
firm_locations
```

```python
locations = extract_detail(firm_locations)
locations = locations.apply(lambda x: x.strip("\n"))
locations
```

```python
locs = locations.apply(lambda x: x.split(", "))
lst = []
for loc in locs:
    lst += loc
    
new_lst = [l.strip() for l in lst]
loctions = pd.Series(new_lst).value_counts(ascending=True)
loctions
```
